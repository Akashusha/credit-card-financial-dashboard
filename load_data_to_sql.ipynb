```python
import pandas as pd
from sqlalchemy import create_engine

# read csv
df = pd.read_csv(
    "C:/Users/akash/Data Projects/Credit_Card_Analysis/credit_card.csv"
)

# mysql connection
engine = create_engine(
    "mysql+mysqlconnector://root:root@localhost:3306/ccdb"
)

# insert into existing table
df.to_sql(
    "cc_detail",
    con=engine,
    if_exists="append",   # VERY IMPORTANT
    index=False,
    chunksize=10000
)
```
